```python
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
from keras.layers import Dropout
from keras import regularizers
```

```python
def create_model(dim, out):
    model = Sequential()
    model.add(Dense(250, input_dim = dim, kernel_initializer='normal', kernel_regularizer=regularizers.l2(0.001), activation='relu'))
#     model.add(Dropout(0.01))
    model.add(Dense(150, kernel_initializer='normal', kernel_regularizer=regularizers.l2(0.001), activation='relu'))
#     model.add(Dropout(0.01))
    
    model.add(Dense(out, activation='softmax'))
    
    adam = Adam(lr=0.001)
    
    model.compile(loss='categorical_crossentropy', optimizer='rmsprop', metrics=['accuracy'])

    return model
```

```python
# model = create_model(300, 96)
model = create_model(300, 91)
print(model.summary())
```
